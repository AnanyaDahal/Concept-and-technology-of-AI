<h2> 3.3 - Exploratory Data Analysis with Group-by Method</h2>


```python
#1)
total_fare = df['Fare'].sum()
print(total_fare)
```

```python
#2)
first_class_passengers = df[df['Pclass'] == 1]
total_fare_first_class = first_class_passengers['Fare'].sum()
print(total_fare_first_class)
```

```python
#3)
# Second class
second_class_passengers = df[df['Pclass'] == 2]
total_fare_second_class = second_class_passengers['Fare'].sum()

# Third class
third_class_passengers = df[df['Pclass'] == 3]
total_fare_third_class = third_class_passengers['Fare'].sum()

print("Total fare for First Class:", total_fare_first_class)
print("Total fare for Second Class:", total_fare_second_class)
```

```python
#4)
fare_totals = [total_fare_first_class, total_fare_second_class, total_fare_third_class]
fare_totals
```

```python
#5)
# Calculate the overall total fare
overall_total_fare = df['Fare'].sum()

# Calculate proportion of fare revenue by class
proportion_first_class = total_fare_first_class / overall_total_fare
proportion_second_class = total_fare_second_class / overall_total_fare
proportion_third_class = total_fare_third_class / overall_total_fare

# Stores in a list
proportions = [proportion_first_class, proportion_second_class, proportion_third_class]

# Print
print("Proportion of fare revenue by class:")
print("First Class:", proportion_first_class)
print("Second Class:", proportion_second_class)
print("Third Class:", proportion_third_class)
```
